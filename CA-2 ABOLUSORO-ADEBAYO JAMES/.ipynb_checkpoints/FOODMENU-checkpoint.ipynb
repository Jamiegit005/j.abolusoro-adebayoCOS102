```python
import tkinter as tk
from tkinter import messagebox

# Define the algorithm for calculating total charges and discount
def calculate_total(name, orders):
    prices = {
        'Juliet Rice': 380,
        'Coconut Fried Rice': 350,
        'Jollof Spaghetti': 350,
        'Set Chi Chicken': 400,
        'Grilled Chicken Wing': 200,
        'Fried Beef': 200,
        'Fried Fish': 200,
        'Sauteed Shrimps': 400,
        'Poundo Yam': 450,
        'Egusi': 480,
        'Atama Soup': 350,
        'Water': 150,
        'Cloud Nine (50cl)': 200,
        'PET Dera': 350,
        'PET Drink (50cl)': 150,
        'Orchard Mist': 900,
        'Fresh Yo': 350,
        'Pineapple Juice': 190,
        'Mango Juice': 350,
        '2000 Drink': 200
    }
    total = sum(prices[item] * quantity for item, quantity in orders.items())
    
    # Apply discounts
    if total < 1000:
        discount = 0
    elif total < 2500:
        discount = total * 0.10
    elif total < 5000:
        discount = total * 0.15
    else:
        discount = total * 0.25
    
    discounted_total = total - discount
    return total, discount, discounted_total

# Define the GUI for inputting orders
def take_order():
    order_window = tk.Toplevel(root)
    order_window.title("Place Your Order")
    
    tk.Label(order_window, text="Name:").grid(row=0, column=0)
    name_entry = tk.Entry(order_window)
    name_entry.grid(row=0, column=1)
    
    menu = {
        'Juliet Rice': tk.IntVar(),
        'Coconut Fried Rice': tk.IntVar(),
        'Jollof Spaghetti': tk.IntVar(),
        'Set Chi Chicken': tk.IntVar(),
        'Grilled Chicken Wing': tk.IntVar(),
        'Fried Beef': tk.IntVar(),
        'Fried Fish': tk.IntVar(),
        'Sauteed Shrimps': tk.IntVar(),
        'Poundo Yam': tk.IntVar(),
        'Egusi': tk.IntVar(),
        'Atama Soup': tk.IntVar(),
        'Water': tk.IntVar(),
        'Cloud Nine (50cl)': tk.IntVar(),
        'PET Dera': tk.IntVar(),
        'PET Drink (50cl)': tk.IntVar(),
        'Orchard Mist': tk.IntVar(),
        'Fresh Yo': tk.IntVar(),
        'Pineapple Juice': tk.IntVar(),
        'Mango Juice': tk.IntVar(),
        '2000 Drink': tk.IntVar()
    }

    # Display menu items
    for index, (item, quantity_var) in enumerate(menu.items()):
        tk.Label(order_window, text=item).grid(row=index+1, column=0)
        tk.Entry(order_window, textvariable=quantity_var).grid(row=index+1, column=1)

    def confirm_order():
        name = name_entry.get()
        orders = {item: quantity_var.get() for item, quantity_var in menu.items() if quantity_var.get() > 0}
        if orders:
            total, discount, discounted_total = calculate_total(name, orders)
            messagebox.showinfo("Order Summary", f"Customer Name: {name}\nTotal: N{total}\nDiscount: N{discount}\nTotal with Discount: N{discounted_total}")
        else:
            messagebox.showinfo("Order Error", "Please add at least one item to your order.")
    
    tk.Button(order_window, text="Confirm Order", command=confirm_order).grid(row=len(menu)+2, column=1)
    
# Main GUI
root = tk.Tk()
root.title("PAU Cafeteria Menu")

menu_button = tk.Button(root, text="Place Your Order", command=take_order)
menu_button.pack(pady=20)

root.mainloop()
```

```python
ALGORITHM
1.	calculate_total(name, orders):
•	This helps calculate the total charges, discount, and discounted total for a given set of orders.
•	It takes the customer's name and a dictionary of orders as input.
•	It defines dictionary “prices” containing the prices of various menu items.
•	It iterates through the orders, multiplying the price of each item by its quantity and summing them up to calculate the total.
•	Based on the total amount, it applies different discount rates (0%, 10%, 15%, or 25%) accordingly.
•	Finally, it returns the total charges, discount amount, and discounted total.
2.	take_order():
•	This function creates a new window using Toplevel() for taking customer orders.
•	It displays input fields for the customer's name and checkboxes for selecting menu items and their quantities.
•	When the customer confirms the order, it extracts the name and selected items with their quantities.
•	If at least one item is selected, it calculates the total charges, discount, and discounted total using calculate_total() and shows a summary message box with this information.
•	If no items are selected, it flags an error.

When you run the code it opens the "pau cafeteria menu" then u press place your order, you then fill in your name and order what ever food u desire
```
